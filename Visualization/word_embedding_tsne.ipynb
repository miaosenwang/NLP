```python
import pylab
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

from sklearn.manifold import TSNE
from sklearn.externals import joblib
```

```python
init_embedding = np.load('init_embedding.npy')

word_order = joblib.load('word_order.pkl') 
```

```python
model_2d = TSNE(n_components=2, random_state=0)
np.set_printoptions(suppress=True)
fitted_2d = model_2d.fit_transform(init_embedding) 
```

```python
model_3d = TSNE(n_components=3, random_state=0)
np.set_printoptions(suppress=True)
fitted_3d = model_3d.fit_transform(init_embedding)
```

```python
fig_3d = pylab.figure(figsize=(80,80))
ax_3d = Axes3D(fig_3d)

for i in range(100):
 ax_3d.scatter(fitted_3d[i,0],fitted_3d[i,1],fitted_3d[i,2],color='b') 
 ax_3d.text(fitted_3d[i,0],fitted_3d[i,1],fitted_3d[i,2],  '%s' % (word_order[i]), size=40, zorder=1,  
 color='k') 

ax_3d.set_zlabel('z')
```

```python
plt.figure(figsize=(80,80))

for label, x, y in zip(word_order[:100], fitted_2d[:100,0], fitted_2d[:100,1]):
    plt.scatter(x, y)
    plt.annotate(label, xy=(x,y), size = 40)
    
plt.show()
```
